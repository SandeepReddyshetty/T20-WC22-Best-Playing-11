```python
import pandas as pd
import json
```

### Match Summary

```python
with open("/content/t20_wc_match_results.json") as f:
  df=json.load(f)
df_match=pd.DataFrame(df[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.info()
```

```python
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
df_match.head()
```

```python
match_id_dict={}

for index,row in df_match.iterrows():
  key1=row['team1'] + ' Vs ' + row['team2']
  key2=row['team2'] + ' Vs ' + row['team1']
  match_id_dict[key1]=row['match_id']
  match_id_dict[key2]=row['match_id']

match_id_dict

```

```python
df_match.to_csv('/content/matchsummary.csv', index=False)
```

### Batting Summary

```python
with open("/content/t20_wc_batting_summary.json") as f:
  df=json.load(f)

  all_records=[]
  for rec in df:
    all_records.extend(rec['battingSummary'])

df_batting=pd.DataFrame(all_records)
df_batting.tail()
```

```python
df_batting['out/not_out']=df_batting.dismissal.apply(lambda x:"out" if len(x)>0 else "not_out")
```

```python
df_batting.head(10)
```

```python
df_batting.drop(columns=['dismissal'],inplace=True)
```

```python
df_batting.head(11)
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('†',''))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('\xa0',''))
df_batting.head(11)
```

```python
df_batting['match_id']=df_batting["match"].map(match_id_dict)
df_batting.head()
```

```python
df_batting.to_csv('/content/battingsummary.csv', index=False)
```

### Bowling Summary

```python
with open("/content/t20_wc_bowling_summary.json") as f:
  df=json.load(f)

  all_records=[]
  for rec in df:
    all_records.extend(rec['bowlingSummary'])

all_records
```
